## Key Observations

- Budget phones like Redmi dominate the market with very high review counts.
- High-priced phones (Apple, Samsung) receive fewer reviews but maintain strong ratings.
- Rating alone is not enough — review count is crucial for trust.
- Mid-range phones offer the best balance of price and customer satisfaction.

```python
import matplotlib.pyplot as plt

top_reviewed = df.sort_values(by="rating_count", ascending=False).head(10)

plt.figure()
plt.barh(top_reviewed["product"], top_reviewed["rating_count"])
plt.xlabel("Number of Reviews")
plt.ylabel("Product")
plt.title("Top 10 Most Reviewed Smartphones on Amazon")
plt.gca().invert_yaxis()
plt.show()
```

### 📊 Insight from Visualization
The visualization clearly shows that budget smartphones dominate customer engagement 
on Amazon. Products with affordable pricing receive significantly higher review counts, 
indicating stronger market penetration and trust among users.
